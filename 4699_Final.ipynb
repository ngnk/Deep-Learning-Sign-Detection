```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
recall_df = pd.read_csv('')
precision_df = pd.read_csv('')
AP50_df = pd.read_csv('')
AP95_df = pd.read_csv('')
```

```python
plt.plot(recall_df['Model'], recall_df['320x320'], marker='o', label='320x320')
plt.plot(recall_df['Model'], recall_df['640x640'], marker='x', label='640x640')

plt.xlabel('Model')
plt.ylabel('Recall')
plt.title('Recall Comparison')
plt.legend()

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid lines for better visualization
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
plt.plot(precision_df['MODEL'], precision_df['320x320'], marker='o', label='320x320')
plt.plot(precision_df['MODEL'], precision_df['640x640'], marker='x', label='640x640')
plt.plot(precision_df['MODEL'], precision_df['1280x1280'], marker='x', label='1280x1280')

plt.xlabel('Model')
plt.ylabel('Precision')
plt.title('Precision Comparison')
plt.legend()

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid lines for better visualization
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
plt.plot(AP50_df['Model'], AP50_df['320x320'], marker='o', label='320x320')
plt.plot(AP50_df['Model'], AP50_df['640x640'], marker='x', label='640x640')
plt.plot(AP50_df['Model'], AP50_df['1280x1280'], marker='x', label='1280x1280')

plt.xlabel('Model')
plt.ylabel('mAP50')
plt.title('mAP50 Comparison')
plt.legend()

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid lines for better visualization
plt.tight_layout()  # Adjust layout
plt.show()
```
